Julia script to simulate coin toss

Toss a coin N times (say N = 10) and count number of Heads. Repeat this experiment for M times (say M = 100). Plot a histogram showing distribution of number of heads in 10 successive coin tosses. 

```julia
using Random
using StatsPlots

```

```julia
# Function to simulate 10 coin tosses and count number of heads
function count_heads(p, N)
    # Generate uniformly distributed random numbers U(0,1)
    # Check if they are less-equal to p
    
    tosses = rand(N) .<= p 
    
    # Count the number of Heads (true values)
    return count(x -> x == true, tosses)
end
```

```julia
# Number of succesive tosses
N = 10

# Number of independent experiments/repeats
M = 10000

# Probability of Head in one coin toss
p = 0.6

results = [count_heads(p, N) for _ in 1:M]

# Plot the histogram of the results
histogram(results, bins=0:1:10, xlabel="Number of Heads in 10 Tosses", ylabel="Frequency", title="Distribution of Heads in 10 Coin Tosses", legend=false)
```
